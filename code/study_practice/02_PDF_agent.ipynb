### 나만의 Agent 쟈비스

```python
# 이번장에 필요한 라이브러리 임포트
import os
import pymupdf
from openai import OpenAI
from dotenv import load_dotenv
import streamlit as st
```

```python
# pdf 파일 불러오기
pdf_file_path = "./data/과정기반 작물모형을 이용한 웹 기반 밀 재배관리 의사결정 지원시스템 설계 및 구축.pdf"
doc = pymupdf.open(pdf_file_path)
```

```python
full_text = ''

# pdf에서 스크리닝하여 텍스트 추출
for page in doc: # 문서 페이지 반복
    text = page.get_text() # 페이지 텍스트 추출
    full_text += text
```

```python
# 파일로 저장하기 위한 객체화
pdf_file_name = os.path.basename(pdf_file_path)
pdf_file_name = os.path.splitext(pdf_file_name)[0] # 확장자 제거

# 파일쓰기
txt_file_path = f"./data/{pdf_file_name}.txt"
with open(txt_file_path, 'w', encoding='utf-8') as f:
    f.write(full_text)
```

```python
# pdf 사이즈를 고려하여 텍스트만 추출하기 위한 사이즈 지정 방법
pdf_file_path = "./data/과정기반 작물모형을 이용한 웹 기반 밀 재배관리 의사결정 지원시스템 설계 및 구축.pdf"
doc = pymupdf.open(pdf_file_path)

header_height = 80
footer_height = 80

full_text = ''
```

```python
for page in doc:
    rect = page.rect # 페이지 크기 가져오기
    
    header = page.get_text(clip=(0, 0, rect.width , header_height))
    footer = page.get_text(clip=(0, rect.height - footer_height, rect.width , rect.height))
    text = page.get_text(clip=(0, header_height, rect.width , rect.height - footer_height))

    full_text += text + '\n------------------------------------\n'

# 파일명만 추출
pdf_file_name = os.path.basename(pdf_file_path)
pdf_file_name = os.path.splitext(pdf_file_name)[0] # 확장자 제거

txt_file_path = f'./data/{pdf_file_name}_with_preprocessing.txt'

with open(txt_file_path, 'w', encoding='utf-8') as f:
    f.write(full_text)
```

```python
# gpt 키를 불러온다.
load_dotenv()
api_key = os.getenv('OPENAI_API_KEY')
```

```python
# ② 주어진 텍스트 파일을 읽어들인다.
file_path = './data/과정기반 작물모형을 이용한 웹 기반 밀 재배관리 의사결정 지원시스템 설계 및 구축.txt'

with open(file_path, 'r', encoding='utf-8') as f:
    txt = f.read()
```

```python
def summarize_txt(file_path: str): # ①
    client = OpenAI(api_key=api_key)

    # ③ 요약을 위한 시스템 프롬프트를 생성한다.
    system_prompt = f'''
    너는 다음 글을 요약하는 봇이다. 아래 글을 읽고, 저자의 문제 인식과 주장을 파악하고, 주요 내용을 요약하라. 

    작성해야 하는 포맷은 다음과 같다. 
    
    # 제목

    ## 저자의 문제 인식 및 주장 (15문장 이내)
    
    ## 저자 소개

    
    =============== 이하 텍스트 ===============

    { txt }
    '''

    print(system_prompt)
    print('=========================================')

        # ④ OpenAI API를 사용하여 요약을 생성한다.
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=0.1,
        messages=[
            {"role": "system", "content": system_prompt},
        ]
    )

    return response.choices[0].message.content
```

```python
if __name__ == '__main__':

    summary = summarize_txt(txt)
    print(summary)

    # ⑤ 요약된 내용을 파일로 저장한다.
    with open('./data/GPT_summary.txt', 'w', encoding='utf-8') as f:
        f.write(summary)
```
