```python
# 라이브러리 임포트
from gpt_functions import get_current_time, tools, get_yf_stock_info
from openai import OpenAI
from dotenv import load_dotenv
import os
import json
```

```python
# 모델 응답 설정
load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")

client = OpenAI(api_key=api_key)

def get_ai_response(messages, tools=None):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        tools=tools,
    )
    return response
```

```python
# 주식 티커 입력
ticker = input("주식 티커를 입력하세요: ").upper()  # 여기에 원하는 주식 티커 입력

# 주식 정보 가져오기
stock_info = get_yf_stock_info(ticker=ticker)
print(f"📊 {ticker} 주식 정보:")
print(stock_info)
print()
```

```python
# 투자 분석을 위한 시스템 메시지
system_message = f"""당신은 전문적인 주식 투자 상담사입니다. 

주어진 주식 정보를 바탕으로 다음과 같은 분석을 제공해주세요:

1. **기본 정보 요약**: 회사명, 현재가, 변동률 등
2. **기술적 분석**: 가격 추세, 거래량, 이동평균 등
3. **투자 위험도**: 변동성, 리스크 요인
4. **투자 추천**: 
   - 매수/매도/홀드 중 어느 것을 추천하는지
   - 추천 이유
   - 투자 시 주의사항
5. **포트폴리오 고려사항**: 다른 주식과의 연관성, 분산투자 관점

분석할 주식: {ticker}
주식 정보: {stock_info}

전문적이면서도 이해하기 쉽게 설명해주세요."""
```

```python
# system_message를 넘겨서 messages객체 생성
messages = [
    {"role": "system", "content": system_message}
]
```

```python
# 투자분석 요청
response = get_ai_response(messages)
analysis = response.choices[0].message.content

print("="*60)
print(f"📈 {ticker} 투자 분석 리포트")
print("="*60)
print(analysis)
print("="*60)
```
